```python
tf_list = []
for i in list(df['주소']):
    print(p.search(i))
    if p.search(i) == None:
        tf_list.append(False)
    else: tf_list.append(True)
```

```python
tf_list
```

```python
num_dong = df[tf_list]
num_dong.head()
```
