```python
def clean_string(st):
    splited = st.split(',')
    for s_st in splited:
        n_st = s_st.replace('[', '').replace(']', '').replace("'", "").strip()
        print(n_st)   
```

```python
li = []
for st in dng:
    splited = st.split(',')
    for s_st in splited:
        n_st = s_st.replace('[', '').replace(']', '').replace("'", "").strip()
        print(n_st)
        li.append(n_st)
```

```python
len(li)
```

```python
merge['동'] = li
merge
```

```python
from collections import Counter
cnt = Counter(li)
cnt
```

```python
less_than_x = []
x = 3

for i in cnt:
    print(i)
    print(cnt[i])
    if cnt[i] <= x: less_than_x.append(i)
```

```python
less_than_x
```

```python
more_than_x = []
for st in dng:
    splited = st.split(',')
    for s_st in splited:
        n_st = s_st.replace('[', '').replace(']', '').replace("'", "").strip()
        print(n_st)
        if n_st not in less_than_x:
            more_than_x.append(n_st)
```

```python
print(len(less_than_x), len(more_than_x))
```
