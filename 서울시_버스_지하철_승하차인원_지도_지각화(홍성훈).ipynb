```python
import pandas as pd
import numpy as np

# 정류장별 승하차 인원 데이터프레임
bus_stop_num_of_people = pd.read_csv('2022년_버스노선별_정류장별_시간대별_승하차_인원_정보(07월).csv', encoding='cp949')

# 야간버스에 해당하는 행들만 추출
night_bus_stop_num_of_people = bus_stop_num_of_people[bus_stop_num_of_people['노선번호'].str.contains('N')]
# 위 코드와 동일한 결과 : night_bus_stop_num_of_people = bus_stop_num_of_people[bus_stop_num_of_people['교통수단타입명']=='서울심야버스']
```

```python
# 정류소 정보를 담고 있는 데이터프레임 
bus_stop_place = pd.read_excel('20220429기준_서울시정류소리스트.xlsx')
```

```python
# 각 정류소별 승하차 인원을 알아야 하기에 merge 함수를 사용해 2개의 데이터프레임을 결합
# inner 조인을 수행했기에 일부 정류소의 경우, 정류소 리스트에 존재하지 않는 데이터는 일부 반영되지 않음. 
bus_stop_information = pd.merge(left = night_bus_stop_num_of_people , right = bus_stop_place, how = "inner", left_on = "표준버스정류장ID", right_on='NODE_ID')
```

```python
# 각 정류소를 기준으로 시간대별 승하차인원을 알고 싶었기에 groupby함수를 사용해서 각 정류소를 기준으로 합계를 구함
# 이 때, 합계의 기준이 되는 칼럼들이 인덱스로 가면 사용하기 어렵기에 as_index 파라미터를 False로 하여 바뀌지 않도록 지정
bus_stop_sum_people = bus_stop_information.groupby(['표준버스정류장ID','사용년월','교통수단타입코드','등록일자','NODE_ID','ARS-ID','좌표X','좌표Y','정류소명'], as_index=False).sum()
bus_stop_sum_people
bus_stop_sum_people.columns
```

# 버스정류장 시각화

```python
from folium.plugins import MarkerCluster
import folium

# 서울지역 위도와 경도
center = [37.541, 126.986]
tiles = ['cartodbpositron', 'Stamen Toner', 'OpenStreetMap']

# folium map 객체 생성
m = folium.Map(
    location = [center[0], center[1]],
    zoom_start = 12,
    tiles = tiles[2]
)

# cluster 객체 생성
marker_cluster = MarkerCluster().add_to(m)

# 위도와 경도를 지정해서 marker를 cluster 객체에 추가
for _, row in bus_stop_sum_people.iterrows():
  folium.Marker([row['좌표Y'], row['좌표X']], icon = folium.Icon(color='red', icon='bus', popup=row['정류소명'])).add_to(marker_cluster)

# html 파일로 생성된 지도를 저장
m.save('심야버스_정류장_위치.html')
```

```python
# 야간버스에 해당하는 'N'번 버스 노선만 제외한 추출
normal_bus_stop_num_of_people = bus_stop_num_of_people[~bus_stop_num_of_people['노선번호'].str.contains('N')]
# 위 결과와 동일한 코드, night_bus_stop_num_of_people = bus_stop_num_of_people[~bus_stop_num_of_people['교통수단타입명']=='서울야간버스']
normal_bus_stop_num_of_people
```

```python
# 각 정류소별 위치는 다른 데이터프레임에 존재하고 있기에 merga함수를 사용해서 데이터프레임들을 결합시켜줌
normal_bus_stop_information = pd.merge(left = normal_bus_stop_num_of_people , right = bus_stop_place, how = "inner", left_on = "표준버스정류장ID", right_on='NODE_ID')
# inner 조인을 수행했기에 일부 정류소의 경우, 정류소 리스트에 존재하지 않는 데이터는 일부 반영되지 않음. 
normal_bus_stop_information
```

```python
# 교통수단타입코드의 경우는 마을버스와 간선버스의 코드가 다르기 때문에 drop ( 처음 한번만 실행)
# normal_bus_stop_information = normal_bus_stop_information.drop(['교통수단타입코드'], axis=1)

# 각 정류소를 기준으로 시간대별 승하차인원을 알고 싶었기에 groupby함수를 사용해서 각 정류소를 기준으로 합계를 구함
# 이 때, 합계의 기준이 되는 칼럼들이 인덱스로 가면 사용하기 어렵기에 as_index 파라미터를 False로 하여 바뀌지 않도록 지정
normal_bus_stop_sum_people = normal_bus_stop_information.groupby(['표준버스정류장ID','사용년월','등록일자','NODE_ID','ARS-ID','좌표X','좌표Y','정류소명'], as_index=False).sum()
normal_bus_stop_sum_people
# normal_bus_stop_sum_people.columns
```

# 지하철역 이용객 시각화

```python
# 각 지하철 역의 시간대별 승하차 인원을 담고 있는 데이터프레임
subway_people = pd.read_csv('서울시 지하철 호선별 역별 시간대별 승하차 인원 정보.csv', encoding='cp949')

# 각 지하철역의 호선과 위도, 경도를 담고 있는 데이터프레임
subway_location = pd.read_csv('서울시 역사마스터 정보.csv', encoding='cp949')
subway_location
```

```python
# 지하철 역의 위치는 다른 데이터프레임에 존재하고 있기에 merga함수를 사용해서 데이터프레임들을 결합시켜줌
subway_information = pd.merge(left = subway_people , right = subway_location, how = "inner", left_on = ["지하철역",'호선명'], right_on=['역사명','호선'])
subway_information
```

```python
# 위에서 만들어준 데이터프레임에서 사용월(측정 월)이 가장 최근인 데이터만 추출
subway_information_recent = subway_information[subway_information['사용월']==202207]
subway_information_recent
```

```python
# 동일한 지하철역을 나타내는 행이 중복이 있기에 groupby 함수를 써서 합계를 구해줌
subway_information_line = subway_information_recent.groupby(['호선명','지하철역','위도','경도'], as_index=False).sum()
subway_information_line
```

```python
# 먼저 1호선에 대한 지하철역 시간대별 인원 데이터만 추출
subway_information_line_1 = subway_information_line[subway_information_line['호선명']=='1호선']

# 역사ID를 기준으로 데이터프레임 정렬 (왜냐하면 역사 ID가 지하철 노선의 정류장 순서일 것이라 생각했기 때문)
subway_information_line_1 = subway_information_line_1.sort_values('역사_ID', ascending=True)
subway_information_line_1 = subway_information_line_1.reset_index(drop=True)

# 각 시간대별 최대 승하차 인원들의 값들을 추출
subway_max_1 = subway_information_line_1.max()

subway_information_line_1
```
